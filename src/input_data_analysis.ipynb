```python
'''
    Extrapolation spin and charge densities from finite-temperature Hartree-Fock data
'''

def formulate_lattice_finite_temperature(input_file, width, height):
    input_data = np.loadtxt(input_file)
    tmp = np.zeros([width, height]) 
    tmp_err = np.zeros([width, height])
    tmp_den = np.zeros([width, height])
    
    for index in range(input_data.shape[0]):
        lx = index % width
        ly = int(index / width)
        tmp[lx, ly] = .5 * (input_data[index, 1] - input_data[index, 2]) 
        tmp_den[lx, ly] = input_data[index, 1] + input_data[index, 2]
    # print(tmp)
    return tmp, tmp_err, tmp_den
```

```python
'''
    Extrapolation spin and charge densities from ground-state Hartree-Fock data
'''

def formulate_lattice_ground_state(input_file, width, height):
    input_data = np.loadtxt(input_file)
    tmp = np.zeros([width, height]); tmp_err = np.zeros([width, height])
    tmp_den = np.zeros([width, height])
    
    for index in range(input_data.shape[0]):
        lx = index % width
        ly = int(index / width)
#         tmp[lx, ly] = (-1) ** (lx + ly) * .5 * (input_data[index, 1] - input_data[index, 2]) 
        tmp[lx, ly] = .5 * (input_data[index, 1] - input_data[index, 2]) 
        tmp_den[lx, ly] = input_data[index, 1] + input_data[index, 2]
    return tmp, tmp_err, tmp_den
```

```python
arrow_size = 20; angle = 30;
arrow_up_color = "darkblue";
arrow_dn_color = "darkred"
```

```python
'''
    Plot real-space spin and charge distribution
'''

def plot_spin_charge(ax, width, height, spin, err):
    x = np.arange(width)
    y = np.zeros(width)
    tmp_spin = np.zeros([width, height])
    rol = 0                     

    ## Plot grids to formulate the lattice. 
    for i in np.arange(height // 2, - height // 2, -1):
        ax.plot([-width // 2, width // 2 + 1], [i, i],  zorder = -1, c = 'black', alpha = 1., linewidth = 3)
    
    for j in np.arange(-width // 2 + 1, width // 2 + 1, 1):
        ax.plot([j, j], [height // 2 + 1, - height // 2], zorder = -1, c = 'black', alpha = 1., linewidth = 3)

    # Plot the spin and density correlation function
    cmap = plt.get_cmap('bwr_r')
#     cmap_reformulated = truncate_colormap(cmap, .15, .85)
    
    for i in range(height):
        tmp = []
        staggered = np.array([(-1)**(j+i%2) for j in range(width)])
        for index in range(width):
            tmp_spin[index, i] = staggered[index] * spin[index, i]
            if abs(spin[index, i]) >= 2. * err[index, i] and abs(spin[index, i]) > 10**-8:
                if tmp_spin[index, i] > 0:
                    tmp.append("lightskyblue")
                else:
                    tmp.append("lightsalmon")
            else:
                tmp.append("white")
                
        pos = ax.scatter(x - width // 2 + 1, y + i - height // 2 + 1, \
                         s = 550, c = tmp, linewidths = 7, edgecolors = 'face',\
                         marker = 's', alpha = 1., zorder = -2)
        
    for index1 in range(spin.shape[0]):
        for index2 in range(spin.shape[1]):
            length = spin[index1, index2] * arrow_size * .1
            dx = length * np.sin(2 * np.pi * angle / 360)
            dy = length * np.cos(2 * np.pi * angle / 360)
            x  = index1 - dx / 2
            y  = index2 - dy / 2
        
            if abs(spin[index1, index2]) >= 2. * err[index1, index2] and abs(spin[index1, index2]) > 10**-8:
                if spin[index1, index2] > 0:
                    arrow_color = arrow_up_color
                else:
                    arrow_color = arrow_dn_color
                 
                ax.arrow(x - width // 2 + 1, y - height // 2 + 1, dx, dy, \
                         head_width=.35, head_length=.35, width=.15, \
                         color=arrow_color, zorder=2, clip_on=False, length_includes_head=False)
    
    ax.set_xlim(- width // 2 + .5, width // 2 + .5)
    ax.set_ylim(- height // 2 + .5, height // 2 + .5)
    ax.tick_params(axis = "both", which = "major", labelsize = 25)
    ax.set_xticks(np.arange(-width // 2 + 2, width // 2 + 1, 2))
    ax.set_yticks(np.arange(-height//2 + 2, height // 2 + 1, 2))
```

```python
'''
    Define the lattice size
'''

Lx = 16
Ly = 6
```

```python
'''
    Convert the spin density to the staggered spin density
'''

def add_staggered_phase(input_array):
    tmp_length = input_array.shape[0]
    output_array = np.zeros([tmp_length])
    for ind in range(tmp_length):
        output_array[ind] = (-1)**ind * input_array[ind]
    return output_array
```

```python
'''
    Spliting a floating number into integer and floating parts
'''

def split_param(tmp):
    '''Split a float number into integer and float parts'''
    tmp1=math.floor(tmp)
    tmp2=tmp-tmp1
    tmp2=np.around(tmp2, decimals=1)
    tmp2=math.floor(10*tmp2)
    return tmp1, tmp2
```

```python
'''
    Compuate chi square using spin-resolved electron densities from QMC and MF solutions
'''

def optimization(input_array, length, inv_temp, effectiveU):
    '''
        Loop over UHF solutions at all temperatures and effective Coulomb interactions 
        to determine the optimal U_eff and beta_eff
    '''
    # Store the chi square results
    chi_tmp=np.zeros([inv_temp.shape[0], effectiveU.shape[0] + 1])
    chi_tmp[:, 0] = inv_temp
    
    for index in range(inv_temp.shape[0]):
        for index_tmp in range(effectiveU.shape[0]):
            
            beta_int, beta_float=split_param(inv_temp[index])
            Uint, Ufloat=split_param(effectiveU[index_tmp])
            filename="Data/MF/PBC/tprime-0p2/rho0p875/p02_optimized/L24W8/Spin_tprime-0p20_U"+str(Uint)+"p"+str(Ufloat)+"_beta"+str(beta_int)+"p"+str(beta_float)+".dat"
#             print(filename)
            
            spin_tmp=np.loadtxt(filename)
            chi_square_tmp=0
            
            for index1 in range(spin_tmp.shape[0]):
                if index1 % length > 1:
                    chi_square_tmp+=((spin_tmp[index1, 1]-input_array[index1, 1]) / input_array[index1, 2])**2 + ((spin_tmp[index1, 2]-input_array[index1, 3]) / input_array[index1, 4])**2
#                 chi_square_tmp+=((spin_tmp[index1, 1]-input_array[index1, 1]) / input_array[index1, 2])**2 + ((spin_tmp[index1, 2]-input_array[index1, 3]) / input_array[index1, 4])**2
            chi_tmp[index, index_tmp + 1] = chi_square_tmp/input_array.shape[0]
            
    return chi_tmp
```
